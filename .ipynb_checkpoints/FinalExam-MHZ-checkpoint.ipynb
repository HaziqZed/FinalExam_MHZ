```python
from math import pi
from math import sqrt
from prettytable import PrettyTable
```

```python
figures = []
```

```python
def cylinderVolume():
    print("\n")
    radius = float(input("Enter the radius of the cylinder: "))
    height = float(input("Enter the height of the cylinder: "))
    cyvol = height * pi * radius**2
    print("\n")
    print(f'The volume of the cylinder is {cyvol:.2f} cubic units')
    print("\n")
    figures.append(["Cylinder Volume",int(radius),int(height),f'{cyvol:.2f}'])
    return 
```

```python
def cylinderSurface():
    print("\n")
    radius = float(input("Enter the radius of the cylinder: "))
    height = float(input("Enter the height of the cylinder: "))
    cysa = (2 * (pi * radius**2)) + ((2 * pi * radius) * height)
    print("\n")
    print(f'The surface area of the cylinder is {cysa:.2f} square units')
    print("\n")
    figures.append(["Cylinder Surface Area",int(radius),int(height),f'{cysa:.2f}'])
    return
```

```python
def coneVolume():
    print("\n")
    radius = float(input("Enter the radius of the cone: "))
    height = float(input("Enter the height of the cone: "))
    covol = pi * (radius**2) * (height/3)
    print("\n")
    print(f'The volume of the cone is {covol:.2f} cubic units')
    print("\n")
    figures.append(["Cone Volume",int(radius),int(height),f'{covol:.2f}'])
    return
```

```python
def coneSurface():
    print("\n")
    radius = float(input("Enter the radius of the cone: "))
    height = float(input("Enter the height of the cone: "))
    cosa = (pi * radius) * (radius + sqrt(radius**2 + height**2))
    print("\n")
    print(f'The surface area of the cone is {cosa:.2f} square units')
    print("\n")
    figures.append(["Cone Surface Area",int(radius),int(height),f'{cosa:.2f}'])
    return
```

```python
def print_table():
    x = PrettyTable()
    x.field_names = (["Measurement","Radius","Height","Value"])
    x.add_row(figures[0])
    x.add_row(figures[1])
    x.add_row(figures[2])
    x.add_row(figures[3])
    x.align["Measurement"] = "r"
    print(x)
```

```python
def save_data():
    filename = input("Enter the filename: ")
    with open(filename, mode='w') as out:
        out.write(str(figures))
```

```python
def main():
    display = {}
    display['1']="Calculate the volume of a cylinder" 
    display['2']="Calculate the surface area of a cylinder"
    display['3']="Calculate the volume of a cone"
    display['4']="Calculate the surface area of a cone"
    display['5']="Print Table"
    display['6']="Save Data"
    print("Welcome to Geometry Calculator!")

    while True: 
        print("     Geometry Calculator")
        print("=============================")
        print("1. Calculate the volume of a cylinder") 
        print("2. Calculate the surface area of a cylinder")
        print("3. Calculate the volume of a cone")
        print("4. Calculate the surface area of a cone")
        print("5. Print Table")
        print("6. Save Data")

        selection = input("Choose option (-1 to quit): ") 
        if selection =='1': 
            cylinderVolume()
        elif selection == '2': 
            cylinderSurface()
        elif selection == '3':
            coneVolume()
        elif selection == '4': 
            coneSurface()
        elif selection =='5':
            print_table()
        elif selection =='6':
            save_data()
        elif selection == '-1':
            print("Thanks for using Geometry Calculator")
            break
```

```python
main()
```
